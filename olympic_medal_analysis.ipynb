### 2-2) 대한민국 역대(1976-2008) 하계 올림픽 메달 획득 내역 확인

```python
df['Country'].unique()
```

```python
df_korea = df[df['Country'] == 'Korea, South'].copy()
```

```python
df_korea.head(1)
```

```python
# 중복되는값 검색
df_dup_tmp = df_korea[df_korea.duplicated(subset=['City', 'Year', 'Sport', 'Discipline', 'Event', 'Country_Code', 'Country','Event_gender', 'Medal'], keep=False)].copy()
df_dup_tmp
```

```python
# 중복값 제거

df_korea_dup_drop = df_korea.drop_duplicates(subset=['City', 'Year', 'Sport', 'Discipline', 'Event', 'Country_Code', 'Country','Event_gender', 'Medal'], keep='last').copy()
df_korea_dup_drop
```

```python
# 피봇테이블 작성

korea_metal_table = df_korea_dup_drop.pivot_table(index=['Year','Medal'],values='City',aggfunc=len)
korea_metal_table
```

```python
korea_metal_table.index.get_level_values(1)
```

```python
korea_metal_table
```

```python
korea_metal_table.sort_index(level=[0, 1],ascending=True, key=lambda x: x.map({'Gold': 1, 'Silver': 2, 'Bronze': 3}),inplace=True)
```

```python
korea_metal_table
```

### 2-3) 1996년 애틀란타 올림픽 총 메달 개수 기준 상위 10개 국가 확인하기

```python
df.head()
```

```python
df_atlanta = df[df['Year'] == 1996].copy()
df_atlanta.head()
```

```python
# 중복값 제거

df_atlanta_dup_drop = df_atlanta.drop_duplicates(subset=['City', 'Year', 'Sport', 'Discipline', 'Event', 'Country_Code', 'Country','Event_gender', 'Medal'], keep='last').copy()
df_atlanta_dup_drop
```

```python
country_list = list(df_atlanta_dup_drop['Country'].unique())
```

```python
country_list
```

```python
df_atlanta_dup_drop_pivot = df_atlanta_dup_drop.pivot_table(index='Country',values='Medal',aggfunc=len)
df_atlanta_dup_drop_pivot
```

```python
df_atlanta_dup_drop_pivot.reset_index(inplace=True)
df_atlanta_dup_drop_pivot
```

```python
df_atlanta_dup_drop_pivot.sort_values(by='Medal', ascending=False, inplace=True)
```

```python
df_atlanta_dup_drop_pivot_top10 = df_atlanta_dup_drop_pivot.head(10)
```

### 2-4) 1996년 애틀란타 올림픽 금매달 개수 기준 상위 10개 국가 확인하기

```python
# 중복 제거된 데이터프레임
df_atlanta_dup_drop.head()
```

```python
df_atlanta_medal_pivot = df_atlanta_dup_drop.pivot_table(index='Country',columns='Medal',aggfunc=len)
df_atlanta_medal_pivot
```

```python
df_atlanta_medal_tmp = df_atlanta_medal_pivot['City'].copy()
df_atlanta_medal_tmp.reset_index(inplace=True)
```

```python
df_atlanta_medal_tmp.fillna(0,inplace=True)
df_atlanta_medal_tmp
```

```python
df_atlanta_medal_tmp = df_atlanta_medal_tmp[['Country','Gold','Silver','Bronze']]
df_atlanta_medal_tmp[['Gold','Silver','Bronze']] = df_atlanta_medal_tmp[['Gold','Silver','Bronze']].astype(int)
```

```python
df_atlanta_medal_result  = df_atlanta_medal_tmp.sort_values(by=['Gold','Silver','Bronze'],ascending=False)
```

```python
df_atlanta_medal_result = df_atlanta_medal_result.head(10)
```

```python
df_atlanta_medal_result
```
