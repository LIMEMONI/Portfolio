## 2. 자동화를 위한 코드

```
- url의 날짜만 변경하면 원하는 기간만큼의 데이터를 얻을 수 있다.
```

````python
date = pd.date_range('2022.01.01', periods=100, freq='D')
date
````

````python
date[0]
````

````python
date[0].strftime('%Y-%m-%d')
````

````python
date[0].strftime('%Y.%m.%d')
````

````python
date[0].strftime('%Y&%m&%d')
````

````python
# 문자열 format

test_string = 'hi im {name}'
test_string.format(name='Sejong')

# dir(test_string) 
````

````python
import time
from tqdm import tqdm

movie_date = []
movie_name = []
movie_point = []

for today in tqdm(date):
    url = 'https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date={date}'
    response = urlopen(url.format(date=today.strftime('%Y%m%d')))
    soup = BeautifulSoup(response, 'html.parser')
    
    end = len(soup.find_all('td','point'))
    movie_date.extend([today for _ in range(0, end)])
    movie_name.extend([soup.find_all('div','tit5')[n].a.text for n in range(0, end)])
    movie_point.extend([soup.find_all('td','point')[n].text for n in range(0, end)])
    
    time.sleep(0.5) # 에러방지
    
````

````python
len(movie_date),len(movie_name),len(movie_point)
````

````python
movie_date[:5],movie_name[:5],movie_point[:5]
````

````python
movie = pd.DataFrame({
    'date':movie_date,
    'name':movie_name,
    'point':movie_point,
})
movie.tail()
````

````python
movie.info() # point가 object 타입
````

````python
movie['point'] = movie['point'].astype(float)
movie['point']
````

````python
# 저장

movie.to_csv(
    '../data/03. naver_movie_data.csv', sep=',', encoding='utf-8'
)
````

## 3. 영화 평점 데이터 정리

````python
import numpy as np
````

````python
movie = pd.read_csv('../data/03. naver_movie_data.csv', index_col=0)
movie
````

- 영화 이름으로 인덱스
- 점수 합산
- 100일간 영화 평점 합산 기준 베스트 워스트 10 선정

````python
#  pivot table 

movie_unique = pd.pivot_table(data=movie, index='name', aggfunc=np.sum)
movie_unique
````

````python
movie_best = movie_unique.sort_values(by='point', ascending=False)
movie_best.head()
````

````python
tmp = movie.query('name == ["씽2게더"]')
tmp
````

````python
# 시각화

import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='Malgun Gothic')
%matplotlib inline
````

````python
plt.figure(figsize=(20, 8))
plt.plot(tmp['date'], tmp['point']) # x축 날짜 y축 점수
plt.title('날 별 평점')
plt.xlabel('날')
plt.ylabel('평점')
plt.xticks(rotation='vertical')
plt.legend(labels=['평점 추이'], loc='best')
plt.grid(True)
plt.show()
````
