### 3. 사용할 함수와 그 함수를 테스트하는 코드를 작성

- 화폐단위 문자형을 입력받아 숫자형으로 반환하는 함수
- 주유소 브랜드를 입력하면 GAS_BRAND 데이터를 참고하여 ID 를 반환하는 함수 
- 주소를 입력받아 구 이름을 반환하는 함수 
- 주소를 입력받아 위도, 경도를 반환하는 함수

```python
# 화폐단위 문자형을 입력받아 숫자형으로 반환하는 함수 (테스트 입력 : ‘1,000’)

def money_to_int(m):
    result = m.replace(',','')
    result = int(result)
    return result
```

```python
test = '1,000'
test
```

```python
money_to_int(test),type(money_to_int(test))
```

```python
# 주유소 브랜드를 입력하면 GAS_BRAND 데이터를 참고하여 ID 를 반환하는 함수 (테스트 입력 : ‘SK에너지’)

def brand_to_id(b):
    cur.execute(f"select id from GAS_BRAND where name = '{b}'")
    result = cur.fetchall()

    return result[0][0]
```

```python
test_brand = 'SK에너지'
brand_to_id(test_brand),type(brand_to_id(test_brand))
```

```python
# 주소를 입력받아 구 이름을 반환하는 함수 (테스트 입력 : ‘서울시 강남구 헌릉로 730’)

def address_to_gu(a):
    result = a.split()[1]
    return result
```

```python
test_address = '서울시 강남구 헌릉로 730'
address_to_gu(test_address),type(address_to_gu(test_address))
```

```python
# 주소를 입력받아 위도, 경도를 반환하는 함수 (테스트 입력 : ‘서울시 강남구 헌릉로 730’)

import googlemaps
gmaps_key = 'AIzaSyDl7USkap1LSpZ7DOPDLtz9d8xQHmdJkU0'
gmaps = googlemaps.Client(key=gmaps_key)

def address_to_coord(a):
    tmp = gmaps.geocode(a, language='ko')[-1]
    lat = tmp.get('geometry')['location']['lat']
    lng = tmp.get('geometry')['location']['lng']
    return [lat,lng]
```

```python
# 리스트 작성

name_list = []
address_list = []
brand_list = []
g_price_list = []
d_price_list = []
self_list = []
wash_list = []
charge_list = []
fix_list = []
conv_list = []
allday_list = []
gu_oil_list = []
lat_list = []
lng_list = []

for gu in tqdm_notebook(gu_names):
    element = driver.find_element(By.ID,"SIGUNGU_NM0") # 구버튼 누르기
    element.send_keys(gu) # 구 정보 send
    time.sleep(3)
    
    # 주유소 개수 확인 작업
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    quantity = len(driver.find_elements(By.CSS_SELECTOR,'#body1 > tr'))
    for i in range(quantity):
        
    
        # 주유소명 클릭해서 정보창 뜨우기
        try:
            driver.find_element(By.CSS_SELECTOR,'#body1 > tr:nth-child({}) > td.rlist > a'.format(i+1)).click()
        except:
            continue
            

        # 주유소 정보 확인하기 
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        # 주유소명
        name = soup.select_one('#os_nm').text
        name_list.append(name)

        # 주소
        address = soup.select_one('#rd_addr').text
        address_list.append(address)

        # 브랜드
        brand = soup.select_one('#poll_div_nm').text
        if brand == '알뜰(ex)':
            brand = '알뜰주유소'
        brand_list.append(brand_to_id(brand))

        # 휘발유 가격 
        g_price = soup.select_one('#b027_p').text
        g_price_list.append(money_to_int(g_price))

        # 경유 가격
        d_price = soup.select_one('#d047_p').text
        d_price_list.append(money_to_int(d_price))

        # 셀프 여부
        self = 0 if soup.select_one('#self_icon') is None else 1
        self_list.append(self)

        # 세차장 여부
        wash = 0 if 'off' in soup.select_one('#cwsh_yn').get('src') else 1
        wash_list.append(wash)

        # 충전소 여부
        charge = 0 if 'off' in soup.select_one('#lpg_yn').get('src') else 1
        charge_list.append(charge)

        # 경정비 여부
        fix = 0 if 'off' in soup.select_one('#maint_yn').get('src') else 1
        fix_list.append(fix)

        # 편의점 여부
        conv = 0 if 'off' in soup.select_one('#cvs_yn').get('src') else 1
        conv_list.append(fix)

        # 24시간 운영 여부
        allday = 0 if 'off' in soup.select_one('#sel24_yn').get('src') else 1
        allday_list.append(allday)
        
        # 구
        gu_oil_list.append(address_to_gu(address))
        
        # 위도/경도
        lat_list.append(address_to_coord(address)[0])
        lng_list.append(address_to_coord(address)[1])
        time.sleep(0.3)
```

### 6. 샘플 위도, 경도 정보를 이용하여 1킬로 이내에 위치한 주유소 정보를 검색

```python
station_sql_df.head(1)
```

- 검색이 안돼서 직접확인
- 37.4955706,127.0292985

```python
from tqdm import tqdm
from haversine import haversine
```

```python

for n in tqdm(range(len(station_sql_df['lat']))):
    dist = haversine([37.4955706,127.0292985], [station_sql_df['lat'][n],station_sql_df['lng'][n]])
    if dist <= 1.00:
        print(f"id: {station_sql_df['id'][n]}/ brand: {station_sql_df['brand'][n]}/ name: {station_sql_df['name'][n]}/ address: {station_sql_df['address'][n]}/ distance: {dist} km")
        print('')
 
```
