```python
import pandas as pd
import numpy as np
```

### 첫번째 변수(PB)

```python
df=pd.read_csv("../cleaning/data_all_pb.csv")
df
```

```python
### 필요한 컬럼만 가져오기
temp = df[["time", "stage", "Lot", "runnum", "recipe", "recipe_step", 'ACTUALROTATIONANGLE', 'ACTUALSTEPDURATION', 'ETCHBEAMCURRENT', 'ETCHGASCHANNEL1READBACK', 
              'ETCHPBNGASREADBACK', 'ETCHSOURCEUSAGE', 'FIXTURETILTANGLE', 'FLOWCOOLFLOWRATE', 'FLOWCOOLPRESSURE', 
              'IONGAUGEPRESSURE', "TTF_FlowCool Pressure Dropped Below Limit", "fault"]].copy()
temp
```

```python
### fault값 확인하기
temp["fault"].value_counts()
```

```python
### 수명주기가 600초(10분) 이내면 fault_labeling 값 1로 지정하기

temp["fault_labeling"] = 0

for i in range(len(temp)) :
    if temp["TTF_FlowCool Pressure Dropped Below Limit"][i] <= 600 :
        temp["fault_labeling"][i] = 1
temp
```

```python
### 600초 이내인 경우가 7,269건
temp["fault_labeling"].value_counts()
```

```python
### 저장하기
temp.to_csv("./M02_preprocessing.csv", index=False)
```
