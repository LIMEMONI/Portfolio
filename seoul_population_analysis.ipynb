```python
import numpy as np
```

```python
# 권역별 인구 pivot_table 생성

df_target_pivot = pd.pivot_table(df_target,index='권역',values=['합계', '세대', '여자', '한국인 계', '등록외국인 계', '65세이상고령자'],aggfunc=np.sum)
df_target_pivot
```

```python
df_target_pivot.sort_values(by='합계', ascending=False, inplace=True)
df_target_pivot
```

```python
df_target_pivot
```

```python
# 비율 추가

add_list = ['고령자비율', '외국인비율', '여성비율']
tmp_list = ['65세이상고령자', '등록외국인 계', '여자']
df_target_pivot[add_list] = df_target_pivot[tmp_list].div(df_target_pivot['합계'], axis=0) * 100
df_target_pivot['세대당인구'] = (df_target_pivot['합계'] - df_target_pivot['등록외국인 계']).div(df_target_pivot['세대'], axis=0)

df_target_pivot
```

```python
df_target_pivot.sort_values(by='외국인비율', inplace=True)
df_target_pivot
```

```python
df_target.head()
```

```python
df_target[['고령자비율', '외국인비율', '여성비율']] = df_target[['65세이상고령자', '등록외국인 계', '여자']].div(df_target['합계'], axis=0) * 100
df_target.head()
```

```python
df_target.sort_values(by='세대당인구', ascending=False, inplace=True)
```
