## 4단계: Data 분석하기(가중 평균 & 분산)
***

### 4-1) 지역대륙별 가준 평균 구하기 (15점)
- 3단계 DataFrame(df)의 국가별 인터넷 사용률(internet_use_rate) 및 인당 소득(income_per_person)을 대륙(region) 또는 지역대륙(sub_region)별로 평균을 합산하여 보기 위해서는 단순 평균이 아닌 국가별 인구수(population)을 가중한 평균을 계산

```python

# groupby 메소드를 이용하여 그룹화한 후, apply 메소드를 이용하여 가중평균을 구함
# weights로 population 열을 지정하여 가중치를 적용함

weighted_ave = df.groupby(['region', 'sub_region']).apply(lambda x: pd.Series(
    {'weighted_ave_internet': np.average(x['internet_use_rate'], weights=x['population']), 
     'weighted_ave_income': np.average(x['income_per_person'], weights=x['population'])}))


weighted_ave
```

```python
# 중국과 인도를 제외한 Eastern Asia, Southern Asia의 데이터를 필터링합니다.
df_asia = df[(df['sub_region'].isin(['Eastern Asia', 'Southern Asia'])) & (
    df['country'] != 'China') & (df['country'] != 'India')]
df_asia
```

```python

# groupby 메소드를 이용하여 sub_region으로 그룹화한 후, apply 메소드를 이용하여 가중평균을 구함
# weights로 population 열을 지정하여 가중치를 적용함

weighted_ave = df_asia.groupby(['region', 'sub_region']).apply(lambda x: pd.Series(
    {'weighted_ave_internet': np.average(x['internet_use_rate'], weights=x['population']), 
     'weighted_ave_income': np.average(x['income_per_person'], weights=x['population'])}))


weighted_ave
```
